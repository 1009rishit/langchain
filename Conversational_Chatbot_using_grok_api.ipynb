```python
from langchain.document_loaders import DirectoryLoader
loader = DirectoryLoader('./test/', glob="./*")
documents = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(documents)
vectorstore = Chroma.from_documents(documents=splits, embedding=embedder)
retriever = vectorstore.as_retriever()
```

```python
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)
```

```python
from google.colab import userdata
llm = ChatGroq(
      temperature=0,
      model="mixtral-8x7b-32768",
      # Optional: Set the API key here if not using an environment variable
      api_key=userdata.get('GROQ_API_KEY')
  )
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory
)
```

```python
def predict(question, history):
    result = qa_chain({"question": question})
    return result['answer']
```

```python
import gradio as gr
iface = gr.ChatInterface(
    predict,
    chatbot=gr.Chatbot(height=300),
    textbox=gr.Textbox(placeholder="Ask me anything!", container=False, scale=7),
    title="Conversational RAG Chatbot",
    description="Ask questions about the documents in the './test/' directory.",
    theme="dark",
    examples=[
        "What topics are covered in the documents?",
        "Can you summarize the main points?",
        "How does this information relate to machine learning?",
    ],
    cache_examples=False,
    retry_btn=None,
    undo_btn="Delete Previous",
    clear_btn="Clear",
)
iface.queue().launch(server_name="0.0.0.0", share=True)
```
